```python
billboard_long = pd.melt(billboard,
        id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
        value_name='rank',
        var_name='week'
       )
billboard_long.head()
```

```python
billboard_long.loc[billboard_long.track == 'Loser']
```

```python
billboard_songs = billboard_long[['year', 'artist', 'track', 'time']]
```

```python
billboard_songs.head()
```

```python
billboard_songs.shape
```

```python
billboard_songs = billboard_songs.drop_duplicates()
```

```python
billboard_songs.shape
```

```python
billboard_songs.head()
```

```python
billboard_songs['id'] = billboard_songs.index
```

```python
billboard_songs.head()
```

```python
billboard_ratings = billboard_long.merge(billboard_songs,
                                        on=['year', 'artist', 'track', 'time'])
```

```python
billboard_ratings.head()
```

```python
billboard_ratings = billboard_ratings[['id', 'date.entered', 'week', 'rank']]
```

```python
billboard_ratings.head()
```

```python
billboard_songs.to_csv('./billboard_songs.csv', index=False)
```

```python
billboard_ratings.to_excel('billboard_ratings.xlsx')
```
