```python
#Selecting the columns of a DataFrame one or more columns
c=["age","gender"]
data[c].head()
```

```python
data.age.head()
```

```python
#Selecting rows of a DataFrame one or more rows
#loc[r] /iloc[ir]/ 
```

```python
l=[1,2]
data.loc[l]
```

```python
l=[1,3]
data.iloc[l]
```

```python
#values of r/c
#at[r,c]/ iat[r,c]/ 
data.at[1,"age"]
```

```python
data.iat[1,1]
```

```python
#Slicing using the [ ] operator
#the slice is performed across the index and not the columns
data[:4]
```
